```python
def fft_denoiser(x, n_components, to_real=True):
    n = len(x)
    
    # compute the fft
    fft = np.fft.fft(x, n)
    
    # compute power spectrum density
    # squared magnitud of each fft coefficient
    PSD = fft * np.conj(fft) / n
    
    # keep high frequencies
    _mask = PSD > n_components
    fft = _mask * fft
    
    # inverse fourier transform
    clean_data = np.fft.ifft(fft)
    
    if to_real:
        clean_data = clean_data.real
    
    return clean_data
```

```python
# Noise removal from the dataset

noise_removed_data = fft_denoiser(df['Close'],0.001,True)
plt.plot(df['Date'],df['Close'])
plt.plot(df['Date'],noise_removed_data)
plt.xlabel('Date')
plt.ylabel('Close value')
plt.legend(['Close','Denoised: 0.001'])
plt.show()
```

```python
# Stationary data checking using Augmented Dickey-Fuller test 
# Ho (Null Hypothesis): The time series data is non-stationary
# H1 (alternate Hypothesis): The time series data is stationary

from statsmodels.tsa.stattools import adfuller
values = df.Close
result = adfuller(values)
print('Augmneted Dickey_fuller Statistic: %f' % result[0])
print('p-value: %f' % result[1])
print('critical values at different levels:')
for k, v in result[4].items():
    print('\t%s: %.3f' % (k, v))
    
# We can see from the p-value, it is clearly greater than 0.05 so we fail to reject the null hypothesis 
# which says the time series data is non-stationary or in other words data is strongly non-stationary.
```

```python
# Forecasting future values using ARIMA

data = df['Close']
length = int(len(df['Close']) * 0.85)
train = data.iloc[:length]
test = data.iloc[length:]

from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import r2_score, mean_squared_error

# ARIMA Model
ar_mod = ARIMA(train, order = (3,2,2))
ar_mod = ar_mod.fit()
print(ar_mod.summary())
```

```python
#Splitting the data for arima model
start = len(train)
end = len(train) + len(test) - 1
y_pred = ar_mod.predict(start = start, end = end, typ = 'levels')
y_pred.index = data.index[start:end+1]
```

```python
plt.figure(figsize = (10, 5))

plt.plot(test)
plt.plot(y_pred)
plt.legend(['Test','Prediction'], loc = 'upper left', fontsize = 15)
plt.show()
```

```python
print("Mean Squared Error:")
print(np.sqrt(mean_squared_error(test, y_pred)))

future_date = pd.date_range(start = '2021-03-31', end = '2021-04-30', freq = 'B' )
y_pred1 = ar_mod.predict(start = len(data), end = len(data) + 23 - 1, typ = 'levels')

#plotting the graphs for training, test, predicted data
plt.figure(figsize = (10, 5))
plt.plot(y_pred)
plt.plot(test)
plt.plot(y_pred1)

plt.title('Actual close values vs predicted values plot')
plt.show()
```

```python
# Smoothing of time series data

from statsmodels.tsa.holtwinters import ExponentialSmoothing

alphas = np.arange(0.01, 1, 0.10)
betas = np.arange(0.01, 1, 0.10)
best_alpha, best_beta, best_mae = des_optimizer(train, alphas, betas)
sm_mod = ExponentialSmoothing(train, trend="add").fit(smoothing_level=best_alpha, smoothing_slope=best_beta)

y_pred = sm_mod.forecast(227)

mae = mean_absolute_error(test, y_pred)
train.plot(legend=True, label="Train", title=f"Double Exponential Smoothing, MAE: {round(mae,2)}")
test.plot(legend=True, label="Test", figsize=(6, 4))
y_pred.plot(legend=True, label="Prediction")
plt.show()
```
